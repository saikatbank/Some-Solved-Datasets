```python
data.columns=['Title', 'Distributor', 'Release date(US)','Budget (millions)', 'Opening weekend(North America)','North America', 'Other territories', 'Worldwide']
```

```python
data
```

```python
data['Title']=data['Title'].astype('string')
data['Distributor']=data['Distributor'].astype('string')
```

```python
data.info()
```

```python
data['Release date(US)']=pd.to_datetime(data['Release date(US)'])
```

```python
data
```

```python
data['Budget (millions)']=data['Budget (millions)'].str.replace(r'\D+','',regex=True).astype('float')
data['Opening weekend(North America)']=data['Opening weekend(North America)'].str.replace(r'\D+','',regex=True).astype('float')
data['North America']=data['North America'].str.replace(r'\D+','',regex=True).astype('float')
data['Other territories']=data['Other territories'].str.replace(r'\D+','',regex=True).astype('float')
data['Worldwide']=data['Worldwide'].str.replace(r'\D+','',regex=True).astype('float')
```

```python
data=data.fillna(0)
```

```python
data
```

```python
data.describe()
```

```python
data.corr()
```

```python
x=data.iloc[:,[3,4,5,6]]
y=data['Worldwide']
```

```python
x
```

```python
type(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
coef=reg.coef_
intercept=reg.intercept_
```

```python
print(coef,intercept)
```

```python
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
